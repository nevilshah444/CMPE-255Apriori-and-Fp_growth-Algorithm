#Introduction to Apriori algorithm:
Apriori Algorithm:-
This algorithm is used for mining the given frequent itemsets and devising association rules as output from a given transactional database. In this algorithm , Two parameters are used frequently:

1) Confidence:A measure of uncertainty associated with each discovered frequent patter.

2) Support: an indication of how frequent items occur in the database

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
#for data preprocessing
from mlxtend.frequent_patterns import apriori, association_rules
#required for performing apriori algorithm
```

Dataset Loading (creating the dataframe)

```python
data= pd.read_csv("train.csv")
data.product_name = ["".join(i.split(",")) for i in data.product_name]
labels= data.order_id.unique()
```

Data Preprocessing (using Transaction Encoder)

```python
Transaction= [data[data.order_id == i]["product_name"].tolist() for i in labels]
encoder = TransactionEncoder().fit(Transaction)
```

```python
Transaction = encoder.transform(Transaction)
Data= pd.DataFrame(Transaction,columns=encoder.columns_)
```

Algorithm Execution

```python
Apriori = apriori(Data, min_support = 0.0045, use_colnames = True)
```

```python
#generating the association rules for Apriori Algorithm
association_rules= association_rules(Apriori, metric = "confidence")
```

```python
print("Association Rules generated by  Apriori Algorithm- \n")
for item in association_rules.iterrows():
  print(f"Rule {item[0] + 1}: {tuple(item[1].antecedents)} -> {tuple(item[1].consequents)}")
```

#Introduction to Fpgrowth Algorithm

FpGrowth Algorithm:- FP-growth is an improved and analysed version of the Apriori Algorithm . It is used as analytical process that derives and matches interesting and frequent associations from a given transactional database.

Working:
1) Calculate the presence of individual items in the dataset

2) With the help of minimum support, find the non-frequent items.

3)Keeping the count of individual occurrences in mind, order the itemsets

4)Create a tree and add the transactions one at a time.

#Importing the necessary libraries

```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
FPgrowth = fpgrowth(Data, min_support = 0.0045, use_colnames=True)

# Apriori = apriori(Data, min_support = 0.0045, use_colnames = True)
# association_rules= association_rules(Apriori, metric = "confidence")
```

```python
association_rules_fp= association_rules(FPgrowth, metric = "confidence")
```

```python
#Displaying the generated association rules
print("Association Rules generated using fpgrowth algorithm:- \n")
for item in association_rules_fp.iterrows():
  print(f"Rule {item[0] + 1}: {tuple(item[1].antecedents)} -> {tuple(item[1].consequents)}")
```
